```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('/content/Quote-Equity-TATACONSUM-EQ-21-03-2023-to-21-03-2024.csv')
df.head()
```

```python
print(df.columns)
```

```python
df.set_index('Date ',inplace = True)
df.drop(['series '],axis=1, inplace=True)
```

```python
df.head()
```

```python
for j in df.columns[:]:
    df[j] = df[j].apply(lambda i: float(i.replace(",", "")) if isinstance(i, str) else i)
```

```python
df.head()
```

```python
sns.clustermap(df)
```

```python
df
```

```python
all_cluster= []
for i in range(1,250):
  model = KMeans(i)
  model.fit(df)
  all_cluster.append(model.inertia_)
```

```python
all_cluster
```

```python
import matplotlib.pyplot as plt
```

```python
sns.set()
plt.plot(range(1,12), all_cluster[0:11])
plt.show()
```

```python
model = KMeans(6)
model.fit(df)
```

```python
identifing_cluster = model.fit_predict(df)
identifing_cluster
```

```python
df['cluster'] = identifing_cluster
df
```
